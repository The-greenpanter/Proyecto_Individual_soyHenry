### Manejando los datos anidados por separado

A contuniacion vamos a separar las columnas en series a las cuales vamos a eliminar los valores NaN  ya que (_luego de varias pruebas nos dimos cuenta que nos dan errores al tratar de desanidar al tratrlos como estructuras JSON_)

```python
removeNac= movies['production_companies']
removeNap= movies['production_countries']
# Revisamos el tipo de dato en cada serie generada para poder tratarlos correctamente
print(removeNac.apply(type).value_counts())
print(removeNap.apply(type).value_counts())

```

Vamos a eliminar los valores NaN pero debemos almacenar las columnas donde estan para que a futuro no nos den problemas al hacer el join

```python
def collect_indices_and_values_series(series):
    """
    Vamos a recopilar los indices y los valores de las series que sean booleanos o float, seguramente asociados a un valor Nan
    o null, asi como Strings mal estructurados.
    
    Argumentos:
    series (pd.Series): Serie a inspeccionar.
    
    Returns:
    Una lista de tuplas de indices valor de la serie analisada.
    """
    # Iniciamos una lista/array/tupla vacia
    results = []

    # Iteramos sobre la serie y evaluamos el tipo de dato
    for idx, value in series.items():
        if isinstance(value, (bool, float)):
            results.append((idx, value))
    
    return results
```

```python

# Aplicamos la funcion
boolean_float_companies = collect_indices_and_values_series(removeNac)
boolean_float_countries = collect_indices_and_values_series(removeNap)

# Imprimirmos los resultados
print("production_companies con bool/float:")
print(boolean_float_companies)

print("\nproduction_countries con bool/float:")
print(boolean_float_countries)
```

Hemos ya tratado las series para entender los valores y garantizar que podamos usar `ast.literal_eval`

```python
# Combinamos las series anteriores a dataframe
df = pd.concat([removeNac, removeNap], axis=1)
# Aplicamos ast.literal_eval a las columnas
df['production_companies'] = df['production_companies'].apply(ast.literal_eval)
df['production_countries'] = df['production_countries'].apply(ast.literal_eval)

print(df.head())
print(df['production_companies'].apply(type).value_counts())
print(df['production_countries'].apply(type).value_counts())
```

```python
def collect_indices_and_values(df, column):
    """
    Recopilamos los indcies y valores del dataframe que tiene bolleanos y floats, como anteriormente
    que pueden estar asociados a NaN o null.
    
    Argumentos:
    df (pd.DataFrame):El data frame con las columnas a analizar.
    column (str): nombre de la columna a ispeccionar.
    
    Returns:
    Lista de tupla con valor indice que son booleanos o floats.
    """
    # Inicia una lista vacia
    results = []

    # Iteracion para hayar los valores
    for idx, value in df[column].items():
        if isinstance(value, (bool, float)):
            results.append((idx, value))
    
    return results
```

```python
# Aplicamos la funcion
boolean_float_pc = collect_indices_and_values(df, 'production_companies')
boolean_float_pcn = collect_indices_and_values(df, 'production_countries')

# Evaluamos el resultado
print("production_companies with bool/float:")
print(boolean_float_pc)
print("\nproduction_countries with bool/float:")
print(boolean_float_pcn)
```
