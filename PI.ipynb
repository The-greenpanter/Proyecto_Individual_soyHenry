## Analisando el anidado

Al observar que dentro de genres hay mas datos y que parecen ser un diccionario pero tambien parecen ser un objeto, vamos a observar que tipo de dato es

```python
print(movies['genres'].head())
print(movies['genres'].apply(type).value_counts())
```


Parece que la columna genres es de tipo str en lugar de una lista de diccionarios. Para poder trabajar con esta columna y desanidar los datos correctamente. Para ello vamos a usar la libreria `import ast`

```python
import ast

# Convierte la columna 'genres' de cadenas a listas de diccionarios
movies['genres'] = movies['genres'].apply(ast.literal_eval)

print(movies['genres'].apply(type).value_counts())
```

```python
# Explota la columna 'genres' para separar los diccionarios en filas individuales
exploded_df = movies.explode('genres')
print(exploded_df.head())
```

```python
# Normaliza los diccionarios en la columna 'genres'
genres_df = pd.json_normalize(exploded_df['genres'])
genres_df
```

```python
# Une los datos normalizados con el DataFrame original
final_df = exploded_df.drop(columns='genres').join(genres_df, lsuffix='_original', rsuffix='_genre')
final_df.head()

```

```python
desanidandodf = genres_df.explode('production_companies')
desanidandodf = genres_df.explode('production_countries')
print(desanidandodf.head())
print(desanidandodf.head())
print(desanidandodf['production_companies'].apply(type).value_counts())
print(desanidandodf['production_countries'].apply(type).value_counts())
```

```python
# Paso 1: Convertir NaN float a listas vacías
def convert_nan_to_empty_list(val):
    if isinstance(val, float) and np.isnan(val):
        return []  # Convertir NaN float a lista vacía
    return val

# # Paso 2: Convertir cadenas a listas usando ast.literal_eval
# def safe_literal_eval(val):
#     if isinstance(val, str):
#         try:
#             evaluated = ast.literal_eval(val)
#             if isinstance(evaluated, (list, dict)):  # Solo aceptamos listas o diccionarios
#                 return (evaluated)
#             return val  # Devolver valor original si no es lista ni diccionario
#         except (ValueError, SyntaxError):
#             return val  # Devolver valor original si no puede ser evaluado
#     return val  # No hacer nada si no es una cadena

# Aplicar las funciones a las columnas deseadas
desanidandodf['production_companies'] = desanidandodf['production_companies'].apply(convert_nan_to_empty_list)


desanidandodf['production_countries'] = desanidandodf['production_countries'].apply(convert_nan_to_empty_list)

# Imprimir los tipos de valores para asegurarnos
print(desanidandodf['production_companies'].apply(type).value_counts())
print(desanidandodf['production_countries'].apply(type).value_counts())
```
