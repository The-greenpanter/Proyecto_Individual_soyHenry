### Manejando los datos anidados por separado

A contuniacion vamos a separar las columnas en series a las cuales vamos a eliminar los valores NaN  ya que (_luego de varias pruebas nos dimos cuenta que nos dan errores al tratar de desanidar al tratrlos como estructuras JSON_)

```python
removeNac= movies['production_companies']
removeNap= movies['production_countries']

print(removeNac.apply(type).value_counts())
print(removeNap.apply(type).value_counts())

```

Vamos a eliminar los valores NaN pero debemos almacenar las columnas donde estan para que a futuro no nos den problemas al hacer el join

```python
def collect_indices_and_values_series(series):
    """
    Collect the indices and values of elements in a Series where the value
    is either of type bool or float.
    
    Args:
    series (pd.Series): The Series to inspect.
    
    Returns:
    list: A list of tuples with (index, value) where the value is either bool or float.
    """
    # Initialize an empty list to store the results
    results = []

    # Iterate through the Series and check each value's type
    for idx, value in series.items():
        if isinstance(value, (bool, float)):
            results.append((idx, value))
    
    return results
```

```python

# Ejemplo de cómo usar la función
boolean_float_companies = collect_indices_and_values_series(removeNac)
boolean_float_countries = collect_indices_and_values_series(removeNap)

# Imprimir los resultados
print("production_companies con bool/float:")
print(boolean_float_companies)

print("\nproduction_countries con bool/float:")
print(boolean_float_countries)
```

```python
def collect_indices_and_values(df, column):
    """
    Collect the indices and values of rows where the specified column contains
    bool or float values.
    
    Args:
    df (pd.DataFrame): The DataFrame containing the column to check.
    column (str): The name of the column to inspect.
    
    Returns:
    list: A list of tuples with (index, value) where the value is either bool or float.
    """
    # Initialize an empty list to store the results
    results = []

    # Iterate through the column and check each value's type
    for idx, value in df[column].items():
        if isinstance(value, (bool, float)):
            results.append((idx, value))
    
    return results
```

```python
# Collect indices and values for both production_companies and production_countries
boolean_float_pc = collect_indices_and_values(df, 'production_companies')
boolean_float_pcn = collect_indices_and_values(df, 'production_countries')

# Print the results
print("production_companies with bool/float:")
print(boolean_float_pc)
print("\nproduction_countries with bool/float:")
print(boolean_float_pcn)
```
